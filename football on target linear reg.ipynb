# Football On Target prediction

### read and cleaning the data

```python
import pandas as pd
df=pd.read_csv("football Data.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df1=df.drop(columns=['Club'])
df1.isnull().sum()
df1
```

### taking the inputs and setting the target

```python
x=pd.DataFrame(df1.loc[:,"Matches_Played":"Shots Per Avg Match"])
y=df1['On Target Per Avg Match']
x
```

## Model Building with Linear Regression

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=42)
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
clf.fit(x_train,y_train)
```

### Predicting the model

```python
y_pred=clf.predict(x_test)
y_pred
```

## Evaluation

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2*100
```

# Visualization of the model

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel('Y tested data')
plt.ylabel('Y predicted data')
plt.title("Goals on target Avg Match")
plt.show()
```
