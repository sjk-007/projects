# ML project using placement data
## Loading the data set 


```python
import pandas as pd
df=pd.read_csv("Placement_Data.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

## Preprocessing the data set 

```python
df1=df.drop(columns='salary',axis=1)
df1
```

## setting x dependent and y independent varaiable

```python
x=df1[["ssc_p","hsc_p","degree_p","etest_p","mba_p"]]
y=df1["status"]
x
```

## spliting the model

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,shuffle=True)
```

## importing the SVM algorithm and fitting the model with the training samples

```python
from sklearn import svm
can=svm.SVC(kernel="linear")
can.fit(x_train,y_train)
```

## Model prediction

```python
y_pred=can.predict(x_test)
y_pred
```

## model Evaluation score

```python
from sklearn import metrics
acc=metrics.accuracy_score(y_test,y_pred)
acc*100
```

# Model Building with KNN

```python
# with knn 

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
```

## Model Prediction for KNN

```python
y_pre=knn.predict(x_test)
y_pre
```

## model Evaluation score

```python
ac=metrics.accuracy_score(y_test,y_pre)
ac*100
```

# Hence we conclude with the improvement of best model as KNN for this dataset as KNN with the help of neighbours value it classify the dataset with more accuracy
