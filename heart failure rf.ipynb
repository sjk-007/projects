 # Death Analysis with Heart Failure

```python

import pandas as pd
df=pd.read_csv("heart_failure.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

## setting the input and target variable

```python
x=pd.DataFrame(df.loc[:,"age":"time"])
y=df["DEATH_EVENT"]
x
```

# Model Creation with Random Forest

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=10)
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
```

## train the model with the data

```python
clf.fit(x_train,y_train)
```

### prediction

```python
y_pred=clf.predict(x_test)
y_pred
```

## Evaluation

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
acc*100
```
